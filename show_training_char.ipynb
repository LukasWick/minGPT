```python
import pandas as pd

import matplotlib.pyplot as plt

def plot(df):
    # Plot the data
    # plt.figure(figsize=(10, 6))
    plt.figure(figsize=(10, 6))
    plt.plot(df['epoch'], df["train_loss"], label="Train Loss")
    plt.ylim(0,3)
    plt.xlabel('Epoch')
    plt.ylabel('Score')
    plt.xscale('log')

    plt.title('Loss Scores Over Epochs')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def plot_from_id(run_id):
    df = pd.read_csv(f'./out/chargpt/{run_id}/train_scores.csv')
    plot(df)
```

```python
df = pd.read_csv('./out/chargpt/train_scores.csv') #context=10
plot(df)
```
