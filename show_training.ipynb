```python
import pandas as pd

import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('../out/multiplier/20260105_165352/train_test_scores.csv')
df = pd.read_csv('../out/adder/20260105_200052/train_test_scores.csv')



# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], 100-df["test_score"], label="Test Score")

plt.xlabel('Epoch')
plt.ylabel('Score')
plt.yscale('log')
plt.xscale('log')

plt.title('Train/Test Scores Over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.figure(figsize=(10, 6))
for col in df.columns:
    if col.lower() != 'epoch':
        plt.plot(df['epoch'], df[col], label=col)

plt.xlabel('Epoch')
plt.ylabel('Score')
plt.xscale('log')

plt.title('Train/Test Scores Over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
